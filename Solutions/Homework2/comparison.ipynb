## Comparison

Let's compare the results with my teammates 

**NOTE**: this will tell who is better, just how much the different predictions differ from one another

```python
import numpy as np
import matplotlib.pyplot as plt

pr_me = np.load('./prediction.npy')
pr_andre = np.load('./prediction_andre.npy')
pr_dorian = np.load('./prediction_dorian.npy')
data = np.load('./prediction-challenge-01-data.npz')['test_x']
```

```python
def get_diff(a, b, data):
    diff = a != b
    diff_data = data[diff]
    diff_a = a[diff]
    diff_b = b[diff]
    
    return diff_a, diff_b, diff_data
```

### Boyan vs Andre

```python
diff_me, diff_andre, diff_data = get_diff(pr_me, pr_andre, data)
print('Number of different guesses: {}'.format(len(diff_data)))
print('Accuracy difference: {}'.format(len(diff_data) / len(data)))
for i in range(10, 15):
    print('Andre: {}; Boyan: {}'.format(diff_andre[i], diff_me[i]))
    plt.figure()
    plt.imshow(diff_data[i][0], cmap='gray')
```

### Boyan vs Dorian

```python
diff_me, diff_dorian, diff_data = get_diff(pr_me, pr_dorian, data)
print('Number of different guesses: {}'.format(len(diff_data)))
print('Accuracy difference: {}'.format(len(diff_data) / len(data)))
for i in range(20, 25):
    print('Dorian: {}; Boyan: {}'.format(diff_dorian[i], diff_me[i]))
    plt.figure()
    plt.imshow(diff_data[i][0], cmap='gray')
```

### Andre vs Dorian

```python
diff_andre, diff_dorian, diff_data = get_diff(pr_andre, pr_dorian, data)
print('Number of different guesses: {}'.format(len(diff_data)))
print('Accuracy difference: {}'.format(len(diff_data) / len(data)))
for i in range(20, 25):
    print('Andre: {}; Dorian: {}'.format([diff_andre[i]], diff_dorian[i]))
    plt.figure()
    plt.imshow(diff_data[i][0], cmap='gray')
```
