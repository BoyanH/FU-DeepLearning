### Comparison

Let's see how different our predictions are

```python
import numpy as np
import matplotlib.pyplot as plt

pr_me = np.load('./prediction.npy')
pr_dorian = np.load('../yu he/prediction.npy')
data = np.load('../yu he/prediction-challenge-02-data.npz')['test_x']
```

```python
def get_diff(a, b, data):
    diff = a != b
    diff_data = data[diff]
    diff_a = a[diff]
    diff_b = b[diff]
    
    return diff_a, diff_b, diff_data
```

####  Boyan vs Dorian

```python
%matplotlib inline

diff_me, diff_dorian, diff_data = get_diff(pr_me, pr_dorian, data)
labels = ['cat', 'dog', 'frog']
print('Number of different guesses: {}'.format(len(diff_data)))
print('Number of test samples: {}'.format(len(data)))
for i in range(50):
    plt.figure()
    plt.title('Dorian: {}; Boyan: {}'.format(labels[diff_dorian[i]], labels[diff_me[i]]))
    plt.imshow(diff_data[i][0])
```

### Difference in accuracy

```python
"Accuracy difference: {:.2f}%".format(100. * ((1. * len(diff_me)) / (1. * len(data))))
```
